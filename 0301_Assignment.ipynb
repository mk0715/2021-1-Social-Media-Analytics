```python
import pandas as pd
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
```

```python
df = pd.read_csv('식당URL.csv')
df
```

```python
df1 = df.query('도시 == "서울"')
df1 = df1.sample(10)

df2 = df.query('도시 == "대전"')
df2 = df2.sample(10)

df3 = df.query('도시 == "부산"')
df3 = df3.sample(10)

# 각 도시마다 10개씩만 추출하여 돌림
df = pd.concat([df1, df2, df3])
df
```

```python
link_list = []

for c, n, u in zip(df['도시'], df['식당이름'], df['url']):
    browser = webdriver.Chrome(ChromeDriverManager().install())
    browser.get(u)
    time.sleep(2)
    try:
        total_page = int(browser.find_element_by_css_selector('#taplc_location_reviews_list_resp_rr_resp_0 > div > div:nth-child(13) > div > div > div > a.pageNum.last').text)
    except:
        total_page = 1
    page_list = list(range(0, total_page*10, 10))
    for p in page_list:
        try:    
            new_u = u.split('Reviews')[0] + 'Reviews-or' + str(p) + u.split('Reviews')[1]
            browser.get(new_u)
            time.sleep(2)
            browser.find_element_by_css_selector('div > div.ui_column.is-9 > div.prw_rup.prw_reviews_text_summary_hsx > div > p > span.taLnk.ulBlueLinks').click()
            time.sleep(2)
            more_links = browser.find_elements_by_css_selector('#taplc_location_reviews_list_resp_rr_resp_0 > div > div > div.review-container')
        except:
            continue
        for m in more_links:
            try:
                review_date = m.find_element_by_css_selector('span.ratingDate').text.split('에')[0]
                ids = m.find_element_by_css_selector('div.info_text.pointer_cursor > div:nth-child(1)').text
                review_rate = float(int(m.find_element_by_css_selector('span.ui_bubble_rating').get_attribute('class')[-2:]) / 10)
                review_text = m.find_element_by_css_selector('div.prw_rup.prw_reviews_text_summary_hsx > div > p').text
                review_text = review_text.replace('\n', ' ')
                link_list.append((c, n, review_date, ids, review_rate, review_text))
            except:
                continue
    browser.quit()
```

```python
link_list
```

```python
df2 = pd.DataFrame(link_list, columns=['지역','식당이름','리뷰일', '리뷰어ID', '평점', '리뷰텍스트'])
df2.to_csv('식당 리뷰.csv', index=False)
df2
```
