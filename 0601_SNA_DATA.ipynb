```python
edges = []
for i in range(len(words)):
    for j in range(i+1, len(words)):
        edges.append((words[i], words[j], word_corr[i,j]))
edges = pd.DataFrame(edges, columns=['word1','word2','weight'])
print(edges.shape)
edges.head()
```

```python
threshold = 0
edges_filtered = edges[edges['weight']>threshold].sort_values(by='weight', ascending=False)
edges_filtered.head()
```
